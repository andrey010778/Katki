```python
df['Тип (каток/лыжная трасса)'].unique()
```

```python
def clean_facility(name):
    name = name.lower()  # приводим к нижнему регистру
    
    if 'каток' in name or 'хоккейная коробка' in name:
        return 'каток'
    elif 'лыжная трасса' in name:
        return 'лыжная трасса'
    else:
        return 'горнолыжный склон'
    

# Применяем функцию к столбцу
df['Тип (каток/лыжная трасса)'] = df['Тип (каток/лыжная трасса)'].apply(clean_facility)
```
